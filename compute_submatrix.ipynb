```python
from itertools import product

basis_exponent = {}
def Mu(x):
    base_field = parent(x)
    degree = log(base_field.order(), base)
    base_gen = base_field.gen() #b, c, d之類的
    M = [list(x * base_gen^j) for j in basis_exponent[degree]]#list就是把每一個元素的多項式的係數寫出來，以升冪的方式
    B = [list(base_gen^j) for j in basis_exponent[degree]]
    return Matrix(M) / Matrix(B)

def generate_basis_exponent(array, n):
    groups = []

    for p, e in array:
        num = (base^n - 1) // (base^(p^e) - 1)
        choices = [Integer(0)] + [Integer(num*i) for i in range(1, Integer(p))]
        groups.append(choices)

    groups = list(reversed(groups))

    basis = [sum(t) for t in product(*groups)]

    return basis

def find_submatrix(genb, gens):
    bn = Mu(genb).nrows()
    sn = Mu(gens).nrows()
    small_matrix = []
    for cut1 in range(bn//sn): #除法變成兩條斜線是因為想要出來的數字是整數
        for cut2 in range(bn//sn):
            small_matrix.append(Mu(genb)[sn*cut1:sn*(cut1+1), sn*cut2:sn*(cut2+1)])
    moment = []
    checked_mat = []
    for i in range(base^sn-1):
        checked_mat.append(Mu(gens^i))
    i0 = zero_matrix(GF(base), sn, sn)
    for M in small_matrix:
        for i in range(base^sn-1):
            if(M == checked_mat[i]):
                if(M == checked_mat[0]):
                    moment.append(base^sn-1)
                else:
                    moment.append(i)
                continue
        if(M == i0):
            moment.append(0)
            continue
        
    for count in range(0, len(moment), bn//sn):
        print(moment[count:count+bn//sn])
```
